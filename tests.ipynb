```python
from scipy.stats import ttest_ind
```

```python
median_gdp = data["Explained by: Log GDP per capita"].median()
group_high_gdp = data[data["Explained by: Log GDP per capita"] > median_gdp]["Ladder score"]
group_low_gdp = data[data["Explained by: Log GDP per capita"] <= median_gdp]["Ladder score"]

t_test, p_value = ttest_ind(group_high_gdp, group_low_gdp, equal_var=True)
print("t-test:", t_test)
print("p-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Odrzucamy hipotezÄ™ zerowÄ…: Åšrednie szczÄ™Å›cie rÃ³Å¼ni siÄ™ istotnie miÄ™dzy grupami.")
else:
    print("Nie odrzucamy hipotezy zerowej: Brak istotnej rÃ³Å¼nicy miÄ™dzy grupami.")
```

```python
median_gdp = data["Explained by: Healthy life expectancy"].median()
group_high_gdp = data[data["Explained by: Healthy life expectancy"] > median_gdp]["Ladder score"]
group_low_gdp = data[data["Explained by: Healthy life expectancy"] <= median_gdp]["Ladder score"]

t_test, p_value = ttest_ind(group_high_gdp, group_low_gdp, equal_var=True)
print("t-test:", t_test)
print("p-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Odrzucamy hipotezÄ™ zerowÄ…: Åšrednie szczÄ™Å›cie rÃ³Å¼ni siÄ™ istotnie miÄ™dzy grupami.")
else:
    print("Nie odrzucamy hipotezy zerowej: Brak istotnej rÃ³Å¼nicy miÄ™dzy grupami.")
```

```python
from scipy.stats import chisquare
import numpy as np
```

```python
generosity = data["Explained by: Generosity"]

observed_frequencies, bins = np.histogram(generosity, bins=5)

expected_frequencies = [sum(observed_frequencies) / len(observed_frequencies)] * len(observed_frequencies)

chi2_stat, p_value = chisquare(observed_frequencies, f_exp=expected_frequencies)

print("Statystyka chi-kwadrat:", chi2_stat)
print("p-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Odrzucamy hipotezÄ™ zerowÄ…: RozkÅ‚ad nie jest zgodny z oczekiwaniami.")
else:
    print("Nie odrzucamy hipotezy zerowej: RozkÅ‚ad jest zgodny z oczekiwaniami.")
```
